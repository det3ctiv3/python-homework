```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4]
y=[10,23,27,50]
plt.plot(x,y)
plt.show()
```

```python
a = np.arange(50)
b = np.random.randint(0,50,50)
size = np.abs(np.random.randn(50)) * 100
plt.scatter(a, b, c=b, s=size)
plt.colorbar()
plt.show()
```

```python
categories = ['Red', 'Green', 'Blue', 'Yellow']
values = [10,15,3,6]
values2 = [2,10,6,3]
x = np.arange(len(categories))
plt.bar(x-0.1, values, color='b', width=0.2)
plt.bar(x+0.1, values2, color='r', width=0.2)
plt.title('Basic Bar Chart')
plt.xlabel('Categories')
plt.xticks(x, categories)
plt.ylabel('Values')
plt.show()
```
